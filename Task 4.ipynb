```python
import numpy as np

# Create a NumPy array with the following values: [1, 2, 3, 4, 5].
arr1 = np.array([1, 2, 3, 4, 5])
print(f'Original array: {arr1}')

# Print the shape of the array using the .shape attribute.
print(f'Shape of original array: {arr1.shape}')

# Reshape the array into a 2D array with 5 rows and 1 column.
arr1 = arr1.reshape((5,1))
print(f'Reshaped array: {arr1}')

# Print the shape of the new array.
print(f'Shape of reshaped array: {arr1.shape}')

# Create a second NumPy array with the following values: [6, 7, 8, 9, 10].
arr2 = np.array([6, 7, 8, 9, 10])
print(f'Second array: {arr2}')

# Concatenate the two arrays together horizontally.
result = np.hstack((arr1,arr2.reshape((-1,1))))
print(f'Concatenated array: {result}')

# Compute the resulting array's mean, median and standard deviation.
mean = np.mean(result)
median = np.median(result)
std_dev = np.std(result)

print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'Standard Deviation: {std_dev}')
```
